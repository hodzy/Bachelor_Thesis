```python
import pandas as pd
```

```python
def clean_row_ids(row):
    row['relation_id'] = row['relation_id'].replace('dict_keys(', '')
    row['relation_id'] = row['relation_id'].replace(')', '')
    row['relation_id'] = row['relation_id'].replace("'", '')
    row['relation_id'] = row['relation_id'].replace("[", '')
    row['relation_id'] = row['relation_id'].replace("]", '')

    return row['relation_id']

def clean_relation_labels(row):
    row['relation_labels'] = row['relation_labels'].replace('dict_values(', '')
    row['relation_labels'] = row['relation_labels'].replace(')', '')
    row['relation_labels'] = row['relation_labels'].replace("'", '')
    row['relation_labels'] = row['relation_labels'].replace("[", '')
    row['relation_labels'] = row['relation_labels'].replace("]", '')
    
    return row['relation_labels']
```

```python
def clean_relation_labels_original(row):
    row['relation labels'] = row['relation labels'].replace('dict_values(', '')
    row['relation labels'] = row['relation labels'].replace(')', '')
    row['relation labels'] = row['relation labels'].replace("'", '')
    row['relation labels'] = row['relation labels'].replace("[", '')
    row['relation labels'] = row['relation labels'].replace("]", '')
    
    return row['relation labels']
```

```python
def combine_relations(row):
    relations = []
    relations_len = 0
    
    relations_obj = str(row['relation_obj_labels']).split(', ')
    if len(relations_obj) == 0:
        return row['relation labels']

    relations_sub = str(row['relation labels']).split(' , ')

    for i in range(max(len(relations_sub), len(relations_obj))):
        if i < len(relations_sub) and relations_sub[i] not in relations:
            relations.append(relations_sub[i].strip())
            
        if i < len(relations_obj) and relations_obj[i].strip() and relations_obj[i] not in relations and len(relations_obj[i]) < 40:
            relations.append(relations_obj[i].strip())
                   
    return relations
```

## Training Data

```python
training_data_obj_1 = pd.read_csv('./3-Relation_Linking_Data/0-preprocess/training_data_find_obj.csv')
training_data_obj_2 = pd.read_csv('./3-Relation_Linking_Data/0-preprocess/training_data_find_obj_2.csv')

training_data_obj = pd.concat([training_data_obj_1, training_data_obj_2])
```

```python
training_data_obj['relation_id'] = training_data_obj.apply(clean_row_ids, axis=1)
training_data_obj['relation_labels'] = training_data_obj.apply(clean_relation_labels, axis=1)
```

```python
training_data_obj.head()
```

```python
training_data_obj = training_data_obj.drop_duplicates(subset=['questions', 'target_relation'])
```

```python
len(training_data_obj)
```

```python
training_data_obj.head()
```

```python
training_data_obj = training_data_obj.rename(columns={'questions': 'question', 'relation_id': 'relation_obj_labels', 'relation_labels': 'relation_obj_ids', 'target_relation': 'target relation id'})
```

```python
training_data_obj 
```

```python
training_data = pd.read_csv('./3-Relation_Linking_Data/1-csv_format/training_data.csv').drop_duplicates()
len(training_data)
```

```python
training_data['relation labels'] = training_data.apply(clean_relation_labels_original, axis= 1)
```

```python
training_data
```

```python
new_training_data = pd.merge(training_data, training_data_obj, how="left", on='question')
```

```python
new_training_data = new_training_data.drop_duplicates(subset=['question', 'target relation id_x', 'target relation id_y'])
```

```python
new_training_data = new_training_data.reset_index(drop=True)
```

```python
new_training_data
```

```python
new_training_data['combined relations'] = new_training_data.apply(combine_relations, axis = 1)
```

```python
new_training_data
```

```python
new_training_data.to_csv('./3-Relation_Linking_Data/2-final_input/training_data.csv', index=False)
```
